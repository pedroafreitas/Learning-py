## Nomes de variáveis

- Camel Case: a partir da segunda palavras, a primeira letra é maiúscula - use para **VARIÁVEIS**.
  - numberOfCollegeGraduates

- Pascal Case: identico ao Camel Case, mas com a primeira letra maiúscula - use para **CLASSES**.
  - NumberOfCollgeGraduates
   
- Snake Case: paalvras separadas por underline - use para **FUNCÕES**.
  - number_of_college_graduates
