```python
"""
Problem Set 0:

Saves two numbers given by the user as 'x'and 'y'
Print out number 'x' raised to the power of 'y'
"""
import numpy

x = input('Enter number \"x":')
y = input('Enter number \"y":')

x = int(x)
y = int(y)
print('X**y = {}'.format(x**y))
print('log(x) = {}'.format(int(numpy.log2(x))))
```
